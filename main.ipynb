```python
import pandas as pd
import pygad
import numpy

# Extração e visualização de dados...
df_raw = pd.read_csv('DSGER.csv', sep=' ', index_col=False)
display(df_raw.head())
# display(df_raw.info())

df = pd.get_dummies(df_raw) # técina one-hot
display(df.head())
# display(df.info())

# Algoritmos Genéticos
function_inputs = df.iloc[0].to_list()
print(function_inputs)

desired_output = 0 # Segundo a modelagem, um score maior ou igual a zero significa que o cliente é um bom pagador

def fitness_func(ga_instance, solution, solution_idx):
    output = numpy.sum(solution*function_inputs)
    fitness = 1.0 / numpy.abs(output - desired_output)
    return fitness

fitness_function = fitness_func

num_generations = 600 # Critério de parada, 600 gerações
num_parents_mating = round(len(function_inputs) * 0.1) # número de pais selecionados para reprodução, entre 10% - 20%

sol_per_pop = round(len(function_inputs) * 0.2) # Número de soluções criadas e evoluídas por geração, entre 10% - 20%
num_genes = len(function_inputs) # 62 genes

# Definição do intervalo de geração de genes
init_range_low = -1 # limite do intervalo inferior
init_range_high = 1 # limite do intervalo superior

parent_selection_type = "rws" # Método da roleta
keep_parents = 1 # 10% de elitismo

crossover_type = "single_point" # crossover de 1 corte

# Mutação de aleatória e com taxa de 0.5%
mutation_type = "random"
mutation_percent_genes = 0.5

ga_instance = pygad.GA(num_generations=num_generations,
                       num_parents_mating=num_parents_mating,
                       fitness_func=fitness_function,
                       sol_per_pop=sol_per_pop,
                       num_genes=num_genes,
                       init_range_low=init_range_low,
                       init_range_high=init_range_high,
                       parent_selection_type=parent_selection_type,
                       keep_parents=keep_parents,
                       crossover_type=crossover_type,
                       mutation_type=mutation_type,
                       mutation_percent_genes=mutation_percent_genes)

ga_instance.run()

solution, solution_fitness, solution_idx = ga_instance.best_solution()
print("Parameters of the best solution : {solution}".format(solution=solution))
print("Fitness value of the best solution = {solution_fitness}".format(solution_fitness=solution_fitness))

prediction = numpy.sum(numpy.array(function_inputs)*solution)
print("Predicted output based on the best solution : {prediction}".format(prediction=prediction))
```
