```python
import pandas as pd
import numpy as np

# Extraindo dados do CSV proposto
df_raw = pd.read_csv('DSGER.csv', sep=' ', index_col=False)

# Função que mostra os intervalos (arbitrários) de variáveis a serem categorizadas
def param_max_min(key):
    print(f'Variável: {key.upper()}')
    max = df_raw[key].max()
    min = df_raw[key].min()
    print(f'I: [{min}, {max}]')
    
    print('Categorização:')
    if(4*min < max):
        print(f'{key}1:[0, {max/4}]; {key}2:({max/4}, {2*max/4}]; {key}3:({2*max/4}, {3*max/4}]; {key}4:({3*max/4}, {max}]')
    elif(4*min == max):
        print(f'{key}1: {max/4}; {key}2: {2*max/4}; {key}3: {3*max/4}; {key}4: {max}')
    elif(2*min == max):
        print(f'{key}1: {min}; {key}2: {max}')
    elif(4*min > max):
        print(f'{key}1:[0, {round(max/3)}]; {key}2:({round(max/3)}, {round(2*max/3)}]; {key}3:({round(2*max/3)}, {round(max)}]')

    print('\n')


# Variáveis que serão categorizadas
variaveis = ['idade', 'valor_credito', 'duracao', 'taxa_parcelamento', 
             'residencia_atual', 'numero_creditos', 'pessoas_dependentes']

for variavel in variaveis:
    param_max_min(variavel)

# Função para categorizar as variáveis de forma dinâmica
def categorizar(chave, valor):
    max = df_raw[chave].max()
    min = df_raw[chave].min()
    if(4*min < max):
        if(0 <= valor <= max/4):
            return f'{chave}1'
        elif(max/4 < valor <= 2*max/4):
            return f'{chave}2'
        elif(2*max/4 < valor <= 3*max/4):
            return f'{chave}3'
        elif(3*max/4 < valor <= max):
            return f'{chave}4'
    elif(4*min == max):
        if(valor == max/4):
            return f'{chave}1'
        elif( valor == 2*max/4):
            return f'{chave}2'
        elif(valor == 3*max/4):
            return f'{chave}3'
        elif(valor == max):
            return f'{chave}4'
    elif(2*min == max):
        if(valor == min):
            return f'{chave}1'
        elif( valor == max):
            return f'{chave}2'
    elif(4*min > max):
        if(0 <= valor <= round(max/3)):
            return f'{chave}1'
        elif(round(max/3) < valor <= round(2*max/3)):
            return f'{chave}2'
        elif(round(2*max/3) < valor <= round(max)):
            return f'{chave}3'

# Aplicando categorização
for variavel in variaveis:
    df_raw.loc[:, variavel] = df_raw[variavel].apply(lambda row: categorizar(variavel, row))

# A variável 'tipo_cliente' determina se ele é um bom(1) ou mal(0) pagador. Valores gerados de forma aleatória. 
# df_raw['tipo_cliente'] = np.random.randint(0, 2, len(df_raw))

df_raw

 # Aplicando técina one-hot. Informações do cliente servidas em uma lista binária.
df = pd.get_dummies(df_raw)
display(df.head())
```

```python

import pandas as pd
import numpy as np
import random

# Passo 1) Extração e visualização de dados
Base_transformada = df
Base2 = df_raw = pd.read_csv('AVALIACAO.csv', sep=' ', index_col=False)

display(Base_transformada)

# Passo 3) Definir parâmetros
num_pessoas_na_base = 100
num_individuals = 10
num_genes = 79
weight_range = (-1, 1)

# Passo 4) Gerar população inicial
population = []
for _ in range(num_individuals):
    chromosome = [random.uniform(weight_range[0], weight_range[1]) for _ in range(num_genes)]
    population.append(chromosome)

# for i in range(num_individuals):
#     print('cromossomo',i,':',population[i]) # primeiro individuo da populacao

#Passo 5) Função Fitness

#Passo 5.1) Criar funcao para calcular o vetor de avaliacao
def calcular_avaliacao(population):
    pesowi=[]
    avaliacao = [] # vetor com cada entrada sendo {um vetor dos resultados dos scores relacionados ao cromossomoi (population[i]) avaliados sobre o banco de dados}
    for i in range(num_individuals):
        for j in range(num_pessoas_na_base): #calcula o score associado ao cromossomo i nas pessoas da Base de dados
            score = np.dot(population[i], Base_transformada.iloc[j]) #score associado ao cromossomo i p.i. Base
            if score>0:
                score=1
            else:
                score=2
            pesowi.append(score)
        avaliacao.append(pesowi)
        pesowi=[]

    print('avaliacao1:',avaliacao[0])
    print('avaliacao2:',avaliacao[1])
    print('avaliacao3:',avaliacao[2])
    return avaliacao

#Passo 5.2) Criar funcao para calcular o Indice de acerto (funcao fitess)

def calcular_Ia(avaliacao): # i é o numero do cromossomo que terá seu Ia calculado
    vetor_fitness=[]
    acerto_bom=0
    acerto_mau=0
    n_bom=0
    n_mau=0
    for i in range(num_individuals):
        for j in range(num_pessoas_na_base):
            if avaliacao[i][j]==1 and Base2['avaliacao'].iloc[j]==1:
                acerto_bom = acerto_bom+1

            if avaliacao[i][j]==2 and Base2['avaliacao'].iloc[j]==2:
                acerto_mau = acerto_mau+1

            if Base2['avaliacao'].iloc[j]==1:
                n_bom = n_bom+1

            if Base2['avaliacao'].iloc[j]==2:
                n_mau = n_mau+1

        Tab = acerto_bom/n_bom
        Tam = acerto_mau/n_mau

        #funcao fitness
        Ia = Tab*Tam

        vetor_fitness.append(Ia)

        print(f'acerto_bom: {acerto_bom},acerto_mau: {acerto_mau}, n_bom: {n_bom},n_mau: {n_mau}','.....Indice de acerto:',round(Ia,2))

        acerto_bom=0
        acerto_mau=0
        n_bom=0
        n_mau=0
        
    return vetor_fitness

# Passo 6) funcao roleta

def selecao_roleta(populacao, fitness, taxa_elitismo=0.1):
    # Calcula o número de indivíduos a serem mantidos por elitismo
    num_elitismo = int(len(populacao) * taxa_elitismo)

    # Ordena a população com base no fitness em ordem decrescente
    populacao_ordenada = [x for _, x in sorted(zip(fitness, populacao), reverse=True)]

    # Seleciona os melhores indivíduos por elitismo
    selecionados = populacao_ordenada[:num_elitismo]

    # Calcula as probabilidades de seleção proporcional ao fitness
    total_fitness = sum(fitness)
    probabilidades = [fit / total_fitness for fit in fitness]

    # Realiza o cruzamento e mutação para formar o restante da nova geração
    nova_geracao = selecionados.copy()
    for _ in range(len(populacao) - num_elitismo):
        # Seleciona dois pais aleatórios com base nas probabilidades
        pai1, pai2 = random.choices(populacao_ordenada, weights=probabilidades, k=2)

        # Realiza o cruzamento (exemplo: ponto de corte no meio)
        ponto_corte = round(len(pai1) // (random.randint(2, 50)))
        filho = pai1[:ponto_corte] + pai2[ponto_corte:] #concatenacao

        # Realiza a mutação (exemplo: inverte um bit aleatório) com chances de 0.5% de mutação
        posicao_mutacao = random.randint(0, len(filho) - 1)
        if random.random() <= 0.0005:
            if filho[posicao_mutacao] == 0:
                filho[posicao_mutacao] = 1
            else:
                filho[posicao_mutacao] = 0

        # Adiciona o filho à nova geração
        nova_geracao.append(filho)

    return nova_geracao

# Passo 7) Programa rodando de fato
geracoes = 0
otimo = []

while geracoes < 50:
    
    if(geracoes == 0):
        pop = population
    else:
        pop = selecionados
        
    avaliacao = calcular_avaliacao(pop)
    vetor_fitness = calcular_Ia(avaliacao)
    print('\n--------------------roleta genetica-------------------\n')
    selecionados = selecao_roleta(pop,vetor_fitness,0.1)
    #selecionados = roleta_genetica(population,vetor_fitness)
    for i in range(num_individuals):
        print('selecionado',i,':',selecionados[i])
#     print('melhor selecionado',':',selecionados[0])
    geracoes += 1
    
    if(geracoes == 50):
        otimo = selecionados[0]
        avaliacao_otima = vetor_fitness[0]
        
print('\n')

otimo_arredondado = [round(valor, 5) for valor in otimo]
colunas = list(Base_transformada.columns.values)

print(f'pesos wi ótimos: {otimo_arredondado}')
print('\n')

print(f'Taxa de acerto ótima: {round(avaliacao_otima * 100)}%')
print('\n')

print('Categoria vs Peso')
for index in range(79):
    print(f'{colunas[index]} <-> {otimo_arredondado[index]}')
```
